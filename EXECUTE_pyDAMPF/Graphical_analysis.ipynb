```python
def findfile(name, path):
    for dirpath, dirname, filename in os.walk(path):
        if name in filename:
            return os.path.join(dirpath, name)
```

```python
direc = os.getcwd()

tempall = np.genfromtxt('tempall.txt')
cases = tempall[:,0]
nperr = int(tempall[1,10])

mascara=np.genfromtxt('mascara_aux.txt')
mask=[]
for i in mascara:
    if i == 1 :
        i = True
    else:
        i=False
    mask.append(i)

cantileversaux=[]
for i in range(len(mascara)):
    cantileversaux.append(i+1)

cantileversaux1=np.array(cantileversaux)
cantilevers=cantileversaux1[mask]
```

```python
os.chdir(direc+'/PARALLELBASIC_0/')

#dominio temporal
tt=[]; zt=[];vt=[];forcet=[];maxforcet=[];vdwt=[];hertzt=[];viscot=[];capt=[];ljt=[];dlvot=[]
#dominio espacial
zc=[]; amp1d=[];fase1d=[];dmind=[];dmaxd=[];fmaxd=[];defld=[];ets1d=[];vts1d=[];pts1d=[];fmedd=[];tcd=[]

for i in cases:
    #print("tdom"+str(int(i))+".dfo")
    filepatht = findfile("tdom"+str(int(i))+".dfo", direc)
    #print(filepatht)
    filepathzc = findfile("zcdom"+str(int(i))+".dfo", direc)
    #print(filepathzc)       
    tdom = np.genfromtxt(open(str(filepatht),"r"))        
    tt.append(tdom[:,0])
    zt.append(tdom[:,1])
    vt.append(tdom[:,2])
    forcet.append(tdom[:,4])
    maxforcet.append(tdom[:,5])
    vdwt.append(tdom[:,6])
    hertzt.append(tdom[:,7])
    viscot.append(tdom[:,8])
    capt.append(tdom [:,9])
    ljt.append(tdom [:,10])
    dlvot.append(tdom [:,11])
    zcdom = np.genfromtxt(open(filepathzc,"r"))
    zc.append(zcdom[:,0])
    amp1d.append(zcdom[:,1])
    fase1d.append(zcdom[:,2])
    dmind.append(zcdom[:,3])
    dmaxd.append(zcdom[:,4])
    fmaxd.append(zcdom[:,5])
    defld.append(zcdom[:,9])
    vts1d.append(zcdom[:,7])
    fmedd.append(zcdom[:,8])
    tcd.append(zcdom[:,10])
    
    if (zcdom[:,6] < 0.9).all():
        ets1daux=np.array([0.])
    else:
        ets1daux=zcdom[:,6]
    ets1d.append(ets1daux)
    #ets1d.append(zcdom[:,6])
    
    #pts1d.append(zcdom[:,11])
    
    if (zcdom[:,6] < 0.9).all():
        pts1daux=np.array([0.])
    else:
        pts1daux=zcdom[:,11]
    pts1d.append(pts1daux)
        
os.chdir(direc)        
```

```python
def amp():
#========Amp 1 vs. zc=================
    
    figuras=tuple()
    for i in range(len(cases)):
        dat = pd.DataFrame(data=[zc[i],amp1d[i] ]).T
        dat.columns = ["zc", "amp1d"]
        dat['legend']= 'Cantilever '+str(cantilevers[i]) 
        fig = px.line(dat,x='zc',y='amp1d',color="legend",color_discrete_sequence=[colores[(cantilevers[i])-1]])
        figuras = figuras+fig.data

    fig1= go.Figure(data=figuras)
    fig1.update_layout(title=('$A (z_c)$'),
                   xaxis_title=('$z_c\; [nm]$'),
                   yaxis_title=('$A \; [nm]$'))
    fig1.write_image(direc+"/PLOTS/Amp_zc.png")
    fig1.show()
```

```python
def asp():
#========Fmax vs. Asp=================
    figuras=tuple()
    Aspn = [0.65,0.75,0.85,0.95]
    for i in range(len(cases)):
        fmaxdasp=[]
        for j in range (len(Asp)):
            aux1 = np.where(amp1d[i] == amp1d[i].flat[np.abs(amp1d[i] - Asp[j]).argmin()])
            fmaxdasp.append(fmaxd[i][aux1][0])
        #print(fmaxdasp)
        dat = pd.DataFrame(data=[Aspn,fmaxdasp]).T
        dat.columns = ["Asp", "fmaxdasp"]
        dat['legend']= 'Cantilever '+str(cantilevers[i])
        #print(dat)
        fig = px.line(dat,x='Asp',y='fmaxdasp',color="legend",color_discrete_sequence=[colores[(cantilevers[i])-1]],markers=True)
        figuras = figuras+fig.data


    fig2= go.Figure(data=figuras)
    fig2.update_layout(title=('$F_{max} (Asp)$'),
                   xaxis_title=('$Asp [nm]$'),
                   yaxis_title=('$F_{max} [nN]$'))
    fig2.write_image(direc+"/PLOTS/Fmax_Asp.png")
    fig2.show()
```
